```python
class Car:
  # self is a reference to the instance of the class
  def  __init__(self, brand, model, year):
    self.brand = brand # Attribute
    self.model = model
    self.year = year

  def display_Info(self):
    # Use f-strings for easier formatting
    print(f"Car: {self.brand} {self.model} {self.year}")

# Create instances of the Car class OUTSIDE the class definition
car1 = Car("Toyota", "corolla", 2022)
car3 = Car("Tesla", "Model S", 2022)

# Call methods on the instances
car1.display_Info()
car3.display_Info() # Corrected from car2 to car3
```

```python
class Employee:
  def __init__(self,Name,company,salary):
    self.Name = Name
    self.company = company
    self.salary = salary
  def display_Info(self):
    print(f"Employee: {self.Name} {self.company} {self.salary}")

Employee1 = Employee("John", "Google", 50000)
Employee2 = Employee("Alice", "Microsoft", 60000)


Employee1.display_Info()
Employee2.display_Info()
```

```python
class Employee:
  def __init__(self,emp_id,name,salary):
    self.emp_id = emp_id
    self.name = name
    self.salary = salary
  def display(self):
    print(f"ID:{self.emp_id}, Name: {self.name}, Salary: {self.salary}")

employees = {
    101 : Employee(101,"John",60000),
    102 : Employee(102,"sohn",6000),
    103 : Employee(103,"Rohan",23000),
}

for emp_id,emp in employees.items():
  emp.display()
```
